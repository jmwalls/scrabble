```python
import glob
import string

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from detector import ontology
```

```python
df = pd.concat([pd.read_pickle(p) for p in glob.glob('data/labelled/0002/*.pkl')])
print(df)
```

```python
fig = plt.figure(figsize=(18, 30))
for i, l in enumerate(list(string.ascii_lowercase)):
    im = np.hstack([d.reshape(64, 60, 3) for d in df.loc[df['label'] == ontology.LABELS_TO_CLASSES[l], 'tile']])
    ax = fig.add_subplot(26, 1, i + 1)
    ax.imshow(im)
    ax.axis('off')
```
